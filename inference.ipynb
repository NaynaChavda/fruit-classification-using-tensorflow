```python
import numpy as np
from keras.models import model_from_json
import operator
import cv2
import sys, os
```

```python
width = 640
height = 480
min_prob_val = 0.65 # 

# Loading the model
json_file = open("model-bw.json", "r")
model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(model_json)
# load weights 
loaded_model.load_weights("model-bw.weights.h5")
print("Loaded model from disk")
 
#Preprossessing function
def prepos(img):
    img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    img = cv2.equalizeHist(img)
    img = img/255
    return img
```

```python
classes = {
    0: 'Apple Braeburn 1',
    1: 'Apple Crimson Snow 1',
    2: 'Apple Red 1',
    3: 'Apple Red 2',
    4: 'Apple Red 3',
    5: 'Apple Red Delicious 1',
    6: 'Banana 1',
    7: 'Banana Lady Finger 1',
    8: 'Cucumber 1',
    9: 'Grape Blue 1',
    10: 'Grape White 2',
    11: 'Grape White 3',
    12: 'Lemon 1',
    13: 'Lemon Meyer 1',
    14: 'Limes 1',
    15: 'Onion Red 1',
    16: 'Onion Red Peeled 1',
    17: 'Orange 1',
    18: 'Pear 1',
    19: 'Pear Forelle 1',
    20: 'Pear Monster 1',
    21: 'Pineapple 1',
    22: 'Pineapple Mini 1',
    23: 'Potato Sweet 1',
    24: 'Potato White 1',
    25: 'Tomato 1',
    26: 'Tomato 2',
    27: 'Tomato 4',
    28: 'Tomato Cherry Red 1'
}
```

```python
cap = cv2.VideoCapture(0) # i have using my laptop camera, if you have a extra camera, you should use this code = cv2.VideoCapture(1)
cap.set(10,width)
cap.set(10,height)
```

```python
while True:
    success, imgOriginal = cap.read()
    img = np.asarray(imgOriginal)
    img = cv2.resize(img,(32,32))
    img = prepos(img)
    cv2.imshow("Processsed Image",img)
    img = img.reshape(1,32,32,1)
    #Printing the each class to the screen
    #cv2.putText(imgOriginal, "CLASSES:", (10, 100), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
    #cv2.putText(imgOriginal, "0 = Apple", (10, 120), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
    #cv2.putText(imgOriginal, "1 = Banana", (10, 140), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
    #cv2.putText(imgOriginal, "2 = Lemon", (10, 160), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
    #cv2.putText(imgOriginal, "3 = Stawberry", (10, 180), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
    #### PREDICT
    # Predict the class
    # If the image is grayscale, convert it to RGB format
    img = np.stack((img,) * 3, axis=-1)

    # Reshape to (1, 32, 32, 3) for batch dimension
    img = img.reshape(1, 32, 32, 3)

    # Predict the class
    prediction = loaded_model.predict(img)
    classIndex = np.argmax(prediction, axis=1)[0]

    # Get the class label from the classes dictionary
    sign = classes[classIndex]

    #print(predictions)
    probVal= np.amax(prediction)
    print(sign,probVal)
 
    if probVal> min_prob_val:
        cv2.putText(imgOriginal,str(sign) + "   "+str(probVal),
                    (50,50),cv2.FONT_HERSHEY_COMPLEX,
                    1,(0,0,255),1)
 
    cv2.imshow("Original Image",imgOriginal)
    
    interrupt = cv2.waitKey(10)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
